```python
junk = ""
with open('books.txt', 'r') as fp:
    junk = fp.readlines()
```

```python
isbn_str = []
for val in junk:
    if "isbn" in val:
        isbn_str.append(val)
```

```python
count = 0
while(count<len(isbn_str)):
    isbn_str[count]= isbn_str[count].split("=")[1]
    count+=1
isbn_str
```
